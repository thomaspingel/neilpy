```python
import neilpy
import numpy as np
from skimage.util import apply_parallel
import matplotlib.pyplot as plt
```

# Inpainting a DTM with missing observations

```python
I, meta = neilpy.imread('ground.tif',force_float=True,fix_nodata=True)
```

```python
# How big is the raster?

np.shape(I)
```

```python
# Use parallel processing to divide up the problem in 100x100 squares, with 30 pixel overlap

J = apply_parallel(neilpy.inpaint_nans_by_springs,I.copy(),100,30)
plt.imshow(J)
plt.show()
```

```python
# How many NaNs now?

np.sum(np.isnan(J))
```

```python
# Write out the result

neilpy.imwrite('ground_inpainted.tif',J,meta)
```
